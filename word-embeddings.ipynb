```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=['the glass of milk',

    'the glass of juice',

    'the cup of tea',

    'I am a good boy',

    'I am a good developer',

    'understand the meaning of words',

    'your videos are good',]
sent
```

```python
voc_size=10000
```

```python
one_hot_repn=[one_hot(words,voc_size) for words in sent]
one_hot_repn
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length=8
embedded_docs=pad_sequences(one_hot_repn,padding='pre',maxlen=sent_length)
```

```python
embedded_docs
```

```python
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,dim))
model.build(input_shape=(None,sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```

```python
model.predict(embedded_docs[0].reshape(1,sent_length))
```
